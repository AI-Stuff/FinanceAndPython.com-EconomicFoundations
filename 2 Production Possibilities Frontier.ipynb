```python
Hamburgers = []
Salads = []

for x in range(0,11):
    Hamburgers.append(x*2)
    Salads.append((10-x)*4)
print(Hamburgers)
print(Salads)
```

```python
import matplotlib.pyplot as plt
plt.plot(Hamburgers,Salads)
plt.show()
```

```python
plt.plot(Hamburgers,Salads)
plt.xlabel("Hamburgers")
plt.ylabel("Salads")
plt.show()
```

```python
Hamburgers = []
Salads = []

for x in range(0,11):
    Hamburgers += [x*4]
    Salads += [(10-x)*2]
print(Hamburgers)
print(Salads)

plt.plot(Hamburgers,Salads)
plt.xlabel("Hamburgers")
plt.ylabel("Salads")
plt.show()
```
